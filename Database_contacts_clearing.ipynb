#### Идеи
Можно разделить датасет так: отдельно строки, где есть все кроме имен; отдельно строки, где есть все. 
Потом было бы круто разделить строки по индустриям (добавить отдельную колонку). И по функциям (добавить отдельную колонку). В идеале вообще еще по размеру компаний (отдельная колонка).
А потом добавить отдельную колонку со ссылками на резюме в базе данных (например, том же дропбоксе). 
И вообще уровень сверх - это сделать так, чтобы очень быстро и легко можно было бы добавлять новых людей в эту базу.

Возможно, для перевода всего на англ пригодится maketrans() - Returns a translation table to be used in translations. Или translate() - Returns a translated string.

```python
# теперь добавляю буквы из Phone в Comments на основном датасете

def find_letters(text):
    res = ''
    for c in text:
        if not c.isnumeric():
            res = text[text.find(c):]
            break
    return res

df['Comments'] = df['Phone'].apply(lambda x: ''.join(find_letters(x)))    # - все верно, добавляет нужное в Comments
df
```

```python
# удаляю буквы и то, что после них, из Phone

def digits_to_remain_in_phone(text):
    res = ''
    for c in text:
        if c.isnumeric():
            res += c
        elif c.isnumeric == False:
            break
    return res


df.Phone = df['Phone'].apply(lambda x: ''.join(digits_to_remain_in_phone(x)))
df
```

```python
# проверка, все ли получилось на основном датасете

test = df[df['Phone'].apply(lambda x: not x.isnumeric())]
test
```
